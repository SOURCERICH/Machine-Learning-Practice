```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline
```

```python
class CSVReader:
    def __init__(self, directory):
        self.directory = directory
        self.data_frames = []
    
    def read_csv_files(self):
        csv_files = [f for f in os.listdir(self.directory) if f.endswith('.csv')]
        
        for csv_file in csv_files:
            df = pd.read_csv(os.path.join(self.directory, csv_file))
            
            df['Category'] = os.path.splitext(csv_file)[0]

            self.data_frames.append(df)
    
    def concatenate_data_frames(self):
        final_dataframe = pd.concat(self.data_frames, ignore_index=True)
        return final_dataframe
```

```python
csv_reader_men = CSVReader('data/Men')
csv_reader_men.read_csv_files()
csv_reader_women = CSVReader('data/Women')
csv_reader_women.read_csv_files()
```

```python
df_men = csv_reader_men.concatenate_data_frames()
df_men['Gender'] = 'Male'
df_men.rename(columns={'product_name' : 'Product Name',
                       'link' : 'Link',
                       ' product_images' : 'Product Image',
                       'price' : 'Price',
                       'details' : 'Details'}, inplace = True)
# df_men.info()
```

```python
df_women = csv_reader_women.concatenate_data_frames()
df_women['Gender'] = 'Female'
df_women.rename(columns={'Product_Name' : 'Product Name',
                       'Product_Image' : 'Product Image',}, inplace = True)
# df_women.info()
```
