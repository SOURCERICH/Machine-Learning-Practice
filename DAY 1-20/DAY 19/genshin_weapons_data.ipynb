```python
type_counts = weapons['type'].value_counts()
five_star_counts = weapons[weapons['rarity'] == '5 Stars']['type'].value_counts()
probabilities = (five_star_counts / type_counts).fillna(0)
print("Probabilities of a given weapon type being 5 Star:\n" ,probabilities)
```

```python
import pandas as pd

ascension_material_columns = ['ascension_material_1', 'ascension_material_2', 'ascension_material_3', 'ascension_material_4']

weapons['ascension_materials_combined'] = weapons[ascension_material_columns].apply(lambda row: ','.join(row.dropna().astype(str)), axis=1)

ascension_materials_df = weapons.groupby('ascension_materials_combined')['weapon_name'].count().reset_index()
ascension_materials_df.columns = ['Ascension Materials', 'Count']

print(ascension_materials_df)
```

```python
print(weapons.nunique())
```

```python
material_1_list = weapons['ascension_material_1'].astype(str).tolist()
material_2_list = weapons['ascension_material_2'].astype(str).tolist()
material_3_list = weapons['ascension_material_3'].astype(str).tolist()
material_4_list = weapons['ascension_material_4'].astype(str).tolist()

def process_ascension_materials(materials):
    for i in range(len(materials)):
        if materials[i] == 'nan':
            materials[i] = ''
        else:
            materials[i] = " ".join(materials[i].split()[:-1])
    return materials

material_1_list = process_ascension_materials(material_1_list)
material_2_list = process_ascension_materials(material_2_list)
material_3_list = process_ascension_materials(material_3_list)
material_4_list = process_ascension_materials(material_4_list)

print(material_1_list)
print(material_2_list)
print(material_3_list)
print(material_4_list)
```

```python
keys = [1, 2, 3, 4]

combined_materials = [item for sublist, key in zip([material_1_list, material_2_list, material_3_list, material_4_list], keys) for item in sublist * key]

print(combined_materials)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(combined_materials, bins=20, edgecolor='k', alpha=0.7)
plt.title('Distribution of Combined and Multiplied Materials')
plt.xlabel('Combined and Multiplied Materials')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
