```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('data/Pokemon.csv', index_col=0, encoding='latin')
```

```python
df.head()
```

```python
sns.lmplot(x="Attack", y="Defense", data=df)
```

```python
sns.lmplot(x='Attack', y='Defense', data=df, fit_reg=False, hue='Stage') 
```

```python
sns.lmplot(x='Attack', y='Defense', data=df, fit_reg=False, hue='Stage')

plt.ylim(0, 200)
plt.xlim(0, 160)
```

```python
plt.figure(figsize=(9,6))
sns.boxplot(data=df)
```

```python
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)
plt.figure(figsize=(9,6))
sns.boxplot(data=stats_df)
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(12,8))
sns.violinplot(x='Type 1', y='Attack', data=df)
```

```python
pkmn_type_colors = ['#78C850',  
                    '#F08030',  
                    '#6890F0',  
                    '#A8B820',  
                    '#A8A878',  
                    '#A040A0',  
                    '#F8D030',  
                    '#E0C068',  
                    '#EE99AC',  
                    '#C03028',  
                    '#F85888',  
                    '#B8A038',  
                    '#705898',  
                    '#98D8D8',  
                    '#7038F8',  
                   ]
```

```python
plt.figure(figsize=(11,6))
sns.violinplot(x='Type 1', y='Attack', data=df, palette=pkmn_type_colors)
```

```python
plt.figure(figsize=(11,6))
sns.swarmplot(x='Type 1', y='Attack', data=df, palette=pkmn_type_colors)
```

```python
plt.figure(figsize=(11,6))

sns.violinplot(x='Type 1',
               y='Attack', 
               data=df, 
               inner=None,
               palette=pkmn_type_colors)
 
sns.swarmplot(x='Type 1', 
              y='Attack', 
              data=df, 
              color='k',
              alpha=0.7)
 

plt.title('Attack by Type')
```

```python
stats_df.head()
```

```python
melted_df = pd.melt(stats_df, 
                    id_vars=["Name", "Type 1", "Type 2"],
                    var_name="Stat") 
melted_df.head()
```

```python
print( stats_df.shape )
print( melted_df.shape )
```

```python
sns.swarmplot(x='Stat', y='value', data=melted_df, 
              hue='Type 1')
```

```python
plt.figure(figsize=(12,8))
 
sns.swarmplot(x='Stat', 
              y='value', 
              data=melted_df, 
              hue='Type 1', 
              dodge=True,
              palette=pkmn_type_colors)
 
plt.ylim(0, 260)
 
plt.legend(bbox_to_anchor=(1, 1), loc=2)
```

```python
plt.figure(figsize=(9,6))
sns.histplot(df, x='Attack',
             kde=True)
```

```python
plt.figure(figsize=(9,6))
sns.countplot(x='Type 1', data=df, palette=pkmn_type_colors)
plt.xticks(rotation=-45)
```

```python
g = sns.catplot(x='Type 1', 
                   y='Attack', 
                   data=df, 
                   hue='Stage',  
                   col='Stage',  
                   kind='swarm') 
 
g.set_xticklabels(rotation=-45)
```

```python
sns.kdeplot(x='Attack', y='Defense', data=df)
```

```python
sns.jointplot(x='Attack', y='Defense', data=df)
```
